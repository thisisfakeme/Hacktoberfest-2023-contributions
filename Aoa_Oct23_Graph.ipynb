```python
# https://www.youtube.com/watch?v=KiB1c8oWxJc
import matplotlib.pyplot as plt
import numpy as np

# create data
x = ['1000', '2000', '3000', '4000', '5000']
strat1 = np.array([37.41, 32.01, 29.424, 28.095, 27.300])
strat2 = np.array([37.24, 31.99, 29.418, 28.093, 27.299])
strat3 = np.array([36.74, 31.80, 29.332, 28.038, 27.262])
strat4 = np.array([37.54, 32.06, 29.445, 28.105, 27.306])

# plot bars in stack manner
s1bar=plt.bar(x, strat1, 0.9, color='r')
s2bar=plt.bar(x, strat2, 0.9, bottom=strat1, color='#44a5c2')
s3bar=plt.bar(x, strat3, 0.9, bottom=strat1+strat2, color='#024b7a')
s4bar=plt.bar(x, strat4, 0.9, bottom=strat1+strat2+strat3, color='g')

plt.bar_label(s1bar, label_type='center', color='white')
plt.bar_label(s2bar, label_type='center', color='white')
plt.bar_label(s3bar, label_type='center', color='white')
plt.bar_label(s4bar, label_type='center', color='white')

plt.xlabel("Size of n")
plt.ylabel("Percentage %")
plt.legend(["Strat 1", "Strat 2", "Strat 3", "Strat 4"])
plt.title("Average Performance of Optimal Strategies k=5000")
plt.show()
```
